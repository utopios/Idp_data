#### 2. Statistiques descriptives

```python
%%bigquery
SELECT
  count(*) AS total_transactions,

  min(date) as date_min,
  max(date) as date_max,

  count(DISTINCT store_number) as nb_magasins_uniques,
  count(DISTINCT item_number) as nb_produits_differents,

  ROUND(SUM(sale_dollars), 2) AS montant_total_ventes,

  ROUND(AVG(sale_dollars), 2) AS vente_moyenne_transaction,

  ROUND(AVG(bottles_sold), 2) AS moyenne_bouteilles_vendues,

  MAX(bottles_sold) AS max_bouteilles_vendues
  FROM `bigquery-public-data.iowa_liquor_sales.sales`;
```

```python
%%bigquery
SELECT
  EXTRACT(YEAR FROM date) AS annee,
  COUNT(*) AS nb_transactions,
  ROUND(SUM(sale_dollars), 2) AS ca_total,
  ROUND(AVG(sale_dollars), 2) AS ca_moyen_transaction,
  SUM(bottles_sold) AS total_bouteilles
FROM `bigquery-public-data.iowa_liquor_sales.sales`
GROUP BY annee
ORDER BY annee;
```

```python
%%bigquery
SELECT
  EXTRACT(MONTH FROM date) AS mois,
  FORMAT_DATE('%B', date) AS nom_mois,
  COUNT(*) AS nb_transactions,
  ROUND(SUM(sale_dollars), 2) AS ca_total,
  ROUND(AVG(sale_dollars), 2) AS ca_moyen
FROM `bigquery-public-data.iowa_liquor_sales.sales`
GROUP BY mois, nom_mois
ORDER BY mois;
```

```python
%%bigquery
SELECT
  category_name,
  COUNT(*) AS nb_transactions,
  ROUND(SUM(sale_dollars), 2) AS ca_total,
  ROUND(AVG(sale_dollars), 2) AS ca_moyen,
  SUM(bottles_sold) AS total_bouteilles,
  COUNT(DISTINCT item_number) AS nb_produits_distincts
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE category_name IS NOT NULL
GROUP BY category_name
ORDER BY ca_total DESC
LIMIT 10;
```

### 3. Identification des anomalies

```python
%%bigquery
SELECT
  COUNT(*) AS nb_anomalies,
  ROUND(SUM(sale_dollars), 2) AS montant_total
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE sale_dollars <= 0;

-- Exemples de transactions
SELECT *
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE sale_dollars <= 0
LIMIT 100;

SELECT
  date,
  store_name,
  city,
  item_description,
  bottles_sold,
  sale_dollars,
  ROUND(sale_dollars / bottles_sold, 2) AS prix_unitaire
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE bottles_sold > 10000
ORDER BY bottles_sold DESC;

SELECT
  date,
  COUNT(*) AS nb_transactions
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE date < '2012-01-01' OR date > CURRENT_DATE()
GROUP BY date
ORDER BY date;
```

## REGRESSION LINEAIRE

#### 1. Prparation des donnes

```python
%%bigquery
CREATE OR REPLACE TABLE `ml_models.iowa_sales_clean` AS
SELECT
  date,
  EXTRACT(YEAR FROM date) AS annee,
  EXTRACT(MONTH FROM date) AS mois,
  EXTRACT(DAYOFWEEK FROM date) AS jour_semaine,
  store_number,
  store_name,
  city,
  county,
  category_name,
  item_description,
  bottle_volume_ml,
  state_bottle_cost,
  state_bottle_retail,
  bottles_sold,
  sale_dollars,
  volume_sold_liters
FROM `bigquery-public-data.iowa_liquor_sales.sales`
WHERE
  sale_dollars BETWEEN 10 AND 10000
  AND bottles_sold BETWEEN 1 AND 1000
  AND date >= '2020-01-01'
  AND sale_dollars IS NOT NULL
  AND bottles_sold IS NOT NULL
  AND state_bottle_retail IS NOT NULL
  AND category_name IS NOT NULL
  AND city IS NOT NULL
LIMIT 500000;
```

### 2. CREATION DU MODEL

```python
%%bigquery

CREATE OR REPLACE MODEL `ml_models.iowa_linear_simple`
OPTIONS(
  model_type='LINEAR_REG',
  input_label_cols=['sale_dollars'],
  data_split_method='AUTO_SPLIT'
) AS
SELECT
  bottles_sold,
  state_bottle_retail,
  sale_dollars
FROM `ml_models.iowa_sales_clean`;
```

```python
%%bigquery
SELECT *
FROM ML.EVALUATE(MODEL `ml_models.iowa_linear_simple`);
```

```python
%%bigquery
SELECT *
FROM ML.WEIGHTS(MODEL `ml_models.iowa_linear_simple`)
ORDER BY ABS(weight) DESC;
```

```python
%%bigquery
SELECT
  bottles_sold,
  state_bottle_retail,
  sale_dollars AS reel,
  predicted_sale_dollars AS predit,
  ABS(sale_dollars - predicted_sale_dollars) AS erreur_absolue,
  ROUND(100 * ABS(sale_dollars - predicted_sale_dollars) / sale_dollars, 2) AS erreur_pct
FROM ML.PREDICT(MODEL `ml_models.iowa_linear_simple`,
  (SELECT * FROM `ml_models.iowa_sales_clean` LIMIT 100));
```
